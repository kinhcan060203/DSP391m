```python
import json
with open('view/song_view_by_week.json', 'r', encoding='utf-8') as f:
    data_track = json.load(f)

l_track=[]
for id in data_track:
    # Tính tổng số lượt phát và số tuần
    total_streams = 0
    week_count = 0

    for entry in data_track[id]:
        streams = entry['streams']
        if streams is not None:
            #print(streams)
            streams = int(streams.replace(',', ''))  # Chuyển đổi chuỗi thành số nguyên
            total_streams += streams
            week_count += 1

    # Tính trung bình
    average_streams = total_streams // week_count if week_count > 0 else 0
    l_track.append({"id_track": id,
                    "streamsperweek": average_streams})
with open(f'view/viewperw_2.json', 'w', encoding='utf-8') as f:
    json.dump(l_track, f, ensure_ascii=False, indent=4)
```

```python
import json

# Đọc file JSON trung tâm (track_base_in4) với mã hóa UTF-8
with open(f'data_popu_final/songs2.json', 'r', encoding='utf-8') as f1:
    track_base_in4 = json.load(f1)

# Đọc file JSON top1_50 với mã hóa UTF-8
with open(f'artist_in4/top1_50.json', 'r', encoding='utf-8') as f2:
    top1_50 = json.load(f2)

# Đọc file JSON top50_end với mã hóa UTF-8
with open(f'artist_in4/top50_end.json', 'r', encoding='utf-8') as f3:
    top50_end = json.load(f3)

# Chuyển top1_50 và top50_end thành dictionary dựa trên id_artist để dễ dàng merge
top1_50_dict = {artist['id_artist']: artist for artist in top1_50}
top50_end_dict = {artist['id_artist']: artist for artist in top50_end}

# Merge dữ liệu vào track_base_in4
for track in track_base_in4:
    id_artist = track['id_artist']
    
    # Nếu id_artist có trong top1_50, thêm thông tin vào
    if id_artist in top1_50_dict:
        track.update(top1_50_dict[id_artist])
    
    # Nếu id_artist có trong top50_end, thêm thông tin vào
    if id_artist in top50_end_dict:
        track.update(top50_end_dict[id_artist])

# Ghi dữ liệu đã merge vào file mới
with open(f'data_popu_final\songs2.json', 'w', encoding='utf-8') as mf:
    json.dump(track_base_in4, mf, ensure_ascii=False, indent=4)

print("Merging complete. Check 'merged_file.json'")
```
